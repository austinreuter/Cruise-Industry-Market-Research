```python
# seeing how the data looks
pro = pd.read_csv('Annual_Financial_Statistics.csv')
pro
```

```python

pro.corr()
```

```python
#since I am going to try to predict gross income is not much of a corralation to annual income it is droped to make a more 
# acurate model
pro.drop(['Annual Percent Change', 'Shares Outstanding (Millions of Shares)'], axis = 1)
```

```python
#determine how the data looks 
pro.plot(kind = 'scatter', x= 'Year', y= 'Gross Profit (Millions)')
plt.show()
```

```python
import os
#Defing X varible the independent varieble and y varible dependent varible 
X = pro.drop('Gross Profit (Millions)', axis = 1)
y = pro['Gross Profit (Millions)'].values.reshape(-1, 1)
```

```python
#splitting data training data to model and test varible
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#import model tools 
from sklearn.linear_model import LogisticRegression, LinearRegression
#define logistic regression
classifier = LogisticRegression()
classifier
```

```python
#intially I tried to plot the results of the linear model and the training data was not the same size test varible
# so I normaliz the arrays
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)

```

```python
#normalize the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
#model linear regression fit model
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
#fitting the logistic regression model
classifier.fit(X_train, y_train)
```

```python
#determine how accurate the models are
r2_lin = model.score(X_test_scaled, y_test_scaled)
r2_log = classifier.score(X_test, y_test)
print(f"score log: {r2_log}")
print(f"score linear: {r2_lin}")
```

```python
#linear model is more accurate so plotting it
plt.plot(model.predict(X_train_scaled), y_train_scaled, label = 'linear model', c= "red")
plt.scatter(model.predict(X_test_scaled), y_test_scaled, label = 'testing data', c = "blue")
plt.legend()
plt.title('gross profit vs 11 indpendent data')
plt.xlabel('independent varibles')
plt.ylabel('gross profit scaled')
plt.show()
plt.savefig('C:/Users/Owner/Desktop/thermo fisher/Neurodiversity_in_the_Workplace_Data_Analysis/model_linear.png')
```
